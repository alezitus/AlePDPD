```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
laplace = list()
n = 200
for x in range(0, n):
    a = list(np.random.laplace(size = 2))
    laplace.append(a)
lap = np.array(laplace)
lap.mean()
```

```python
x = lap[:,0]
y = lap[:,1]
plt.plot(x,y)
```

```python
A = np.random.rand(2,2)
S = np.array([x, y])
X = A.dot(S)
x1 = X[0]
x2 = X[1]
plt.figure()
plt.plot(x1)
plt.figure()
plt.plot(x2)
```
