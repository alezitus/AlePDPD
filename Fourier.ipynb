```python
x = np.arange(-0.25, 0.25, 0.001)
f0 = 4
b = 0
y = 1/2
for n in range (1,1000,2):  
    b = 2/(np.pi*n)
    y = y+b*np.sin(2*np.pi*n*x*f0)
plt.plot(x,y)
```

```python
x = np.arange(-0.5, 0.5, 0.001)
L = 1/8
b = 0
y = 0
for n in range (1,3):  
    b = (2*L/(n*np.pi))*((-1)**(n+1))
    y = y+b*np.sin((np.pi*n*x)/L)
plt.plot(x,y)
```
