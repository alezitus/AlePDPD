```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
b0 = np.random.rand(2)
```

```python
cov = np.array([[3, 1], [1,2]])
def mp(cov, num):
    for n in range(num):
        if n == 0:
            bk = cov.dot(b0)
        else:
            bk = cov.dot(bk)
        norm = np.linalg.norm(bk)
        bknorm = bk/norm
    return bknorm
```

```python
def online(a):
    global x
    global W
    global sizex
    global cov
    a = a
    Want = W
    cont = it = 0
    ww = list()
    for n in range(sizex):
        y = (W).dot(x[n,:])
        c = (x[n,:].T)*(y)
        d = y*y*(W)
        dw = a*(c - d)
        Want = W
        W= W+dw
        dif1 = np.abs(W[0][0] - Want[0][0])
        dif2 = np.abs(W[0][1] - Want[0][1])
        if dif1 < a*1.5 and dif2 < a*1.5:
            it += 1
        else:
            it = 0
        #if it == 15 :
            #break
        cont = cont+1
        ww.append(W[0])
    plt.figure()
    plt.title('Online Learning')
    plt.plot(ww)
    #plt.plot(w2)
    print(f'Método: Online Learning\nTaxa de Aprendizagem: {a}\nNúmero de Amostras em X: {sizex}\nVetor W: {W}\nAutovetor: {mp(cov,10)}\nNúmero de iterações até convergência:{cont}')
    print()
```

```python
def batch(a):
    global x
    global W
    global sizex
    global cov
    a = a
    Want = W
    cont = it = 0
    ww = list()
    for n in range(0,sizex,10):
        z1 = list()
        z2 = list()
        z = [m for m in x[n:n+10,:]]
        for v in range (0,10):
            z1.append(z[v][0])
            z2.append(z[v][1])
        z1 = np.array(z1)
        z2 = np.array(z2)
        z1 = z1.mean()
        z2 = z2.mean()
        z = np.array([z1, z2])
        y = (W).dot(z)
        c = (z.T)*(y)
        d = y*y*(W)
        dw = a*(c - d)
        Want = W
        W= W+dw
        dif1 = np.abs(W[0][0] - Want[0][0])
        dif2 = np.abs(W[0][1] - Want[0][1])
        if dif1 < a*1.5 and dif2 < a*1.5:
            it += 1
        else:
            it = 0
        #if it == 5 :
            #break
        cont = cont+1
        ww.append(W[0])
    plt.figure()
    plt.title('Batch Learning')
    plt.plot(ww)
    print(f'Método: Batch Learning\nTaxa de Aprendizagem : {a}\nNúmero de Amostras em X: {sizex}\nVetor W: {W}\nAutovetor: {mp(cov,10)}\nNúmero de iterações até convergência:{cont}')
    print()
```

```python
def batch1(a):
    global x
    global W
    global sizex
    global cov
    a = a
    Want = W
    cont = it = 0
    ww = list()
    w1 = list()
    w2 = list()
    for n in range(0,sizex,10):
        z = list()
        n = list()
        z = [m for m in x[n:n+10,:]]
        for v in range (0,10):
            zz.append(z[v])
        for m in range(len(x)):
            for b in zz:
                soma += b[m]
            media = soma/len(x)
            n.append(media)
        n = np.array(n)
        y = (W).dot(n)
        c = (n.T)*(y)
        d = y*y*(W)
        dw = a*(c - d)
        Want = W
        W= W+dw
        dif1 = np.abs(W[0][0] - Want[0][0])
        dif2 = np.abs(W[0][1] - Want[0][1])
        #if dif1 < a*1.5 and dif2 < a*1.5:
            #it += 1
        #else:
            #it = 0
        #if it == 5 :
            #break
        cont = cont+1
        w1.append(W[0][0])
        w2.append(W[0][1])
        ww.append(W[0])
    plt.figure()
    plt.title('Batch Learning')
    plt.plot(ww)
    print(f'Método: Batch Learning\nTaxa de Aprendizagem : {a}\nNúmero de Amostras em X: {sizex}\nVetor W: {W}\nAutovetor: {mp(cov,10)}\nNúmero de iterações até convergência:{cont}')
    print()
```

```python
W = np.random.rand(1,2)
sizex = 1000
cov = np.array([[3, 1], [1,2]])
mean = np.zeros(2)
b0 = np.random.rand(2)
x = np.random.multivariate_normal(mean = mean, cov = cov, size =sizex)
```

```python
online(0.001)
batch(0.001)
batch1(0.001)
```
